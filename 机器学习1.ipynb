```python
#numpy基础
import numpy as np
print(np.array([1,2,3])) #创建一维数组
print(np.array([(1,2,3),(4,5,6)])) #创建二维数组
print(np.zeros((3,4))) #创建3行4列的二维数组
print(np.ones((2,3,4)))#创建2层3行4列的三维数组
print(np.full((3,4),2)) #创建值2,3行4列的2维数组
print(np.arange(6) )#生成一维数组
print(np.arange(6).reshape(2,3)) #生成一维数组并重塑成2行3列
print(np.eye(3)) #创建3行3列的单位矩阵
print(np.random.rand(2,3) )#生成2行3列随机数组
print(np.random.randint(6,size=(2,3)) )#生成<6的2行3列随机整数数组

#numpy的优点:可以直接进行矩阵运算
myarray1 = numpy.array([2, 2, 2])
myarray2 = numpy.array([3, 3, 3])
print("Addition: %s"% (myarray1 + myarray2))
print("Multiplication: %s" % (myarray1 * myarray2))
```

```python
#matplotlib基础
#import sys
#!{sys.executable} -m pip install basemap #下载basemap库
#conda install basemap-data-hires #下载basemap数据集文件

#用matplotlib画个世界地图
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(projection='mill', llcrnrlat=-90, urcrnrlat=90,  # 创建Basemap对象并设置地图参数
            llcrnrlon=-180, urcrnrlon=180, resolution='c')
m.drawcoastlines()  # 绘制海岸线和国家边界线
m.drawcountries()
plt.show() #显示

#折线图
import matplotlib.pyplot as plt
import numpy
myarray = numpy.array([1, 2, 3]) #这里默认将1,2,3作为y轴坐标,而x轴坐标为0,1,2
plt.plot(myarray)
plt.xlabel('some x axis')
plt.ylabel('some y axis')
plt.show()

#散点图
import matplotlib.pyplot as plt
import numpy
x = numpy.array([1, 2, 3])
y = numpy.array([2, 4, 6])
plt.scatter(x,y)
plt.xlabel('some x axis')
plt.ylabel('some y axis')
plt.show()
```

```python
#pandas基础
import numpy
import pandas
print('一维数据:')
myarray = numpy.array([1, 2, 3])
rownames = ['a', 'b', 'c']
myseries = pandas.Series(myarray, index=rownames) #前数据后索引
print(myseries)
print(myseries[0])
print(myseries['a']) #索引的作用同上

print('二维数据:')
myarray = numpy.array([[1, 2, 3], [4, 5, 6]])
rownames = ['a', 'b']
colnames = ['one', 'two', 'three']
mydataframe = pandas.DataFrame(myarray, index=rownames, columns=colnames) #二维数组
print(mydataframe)
print("method 1:")
print("one column: %s"% mydataframe['one'])
print("method 2:")
print("one column: %s" % mydataframe.one)
print('精确索引:%s'% mydataframe.loc['a','one']) #用.loc实现精确索引
```

```python
#机器学习数据
import csv  #用csv库读取csv文件（csv文件的优点是小,且不必严格行列）
import numpy
filename = 'iris.csv'  #把csv文件上传到jupyter然后读取
raw_data = open(filename, 'r')
print("1.原始数据raw_data:",raw_data)
reader = csv.reader(raw_data, delimiter=',', quoting=csv.QUOTE_NONE)
print("2.读取reader:",reader)
x = list(reader)
print("3.列表情况x:",x)
data = numpy.array(x)
print("行列情况:",data.shape)  #shape输出行、列情况
print(data)

```
